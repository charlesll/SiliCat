```julia
using PyPlot
using Ipopt
using JuMP
```

```julia
data = readcsv("/Users/charles/Dropbox/SiliCat/data/viscosity/viscosity.csv",skipstart=1)
sio2  = data[:,2]./100
tio2  = data[:,3]./100
al2o3 = data[:,4]./100
feo   = data[:,5]./100
mno   = data[:,6]./100
bao   = data[:,7]./100
sro   = data[:,8]./100
mgo   = data[:,9]./100
cao   = data[:,10]./100
li2o  = data[:,11]./100
na2o  = data[:,12]./100
k2o   = data[:,13]./100
h2o   = data[:,14]./100
T     = data[:,15]
n     = data[:,16]
println("Done")

nb_at = [3,3,5,2,2,2,2,2,2,3,3,3,3]

ref = Array{Float64}(data[data[:,2] .== 100,2:16])
data_fit = Array{Float64}(data[data[:,2] .!= 100,2:16])
```

```julia
## Model
mod = Model(solver=IpoptSolver(acceptable_tol = 1e-10))

# starting values

@variable(mod,ni[i=1:13], start = 1.0)
@variable(mod,K[i=1:13] , start = 1.0)
@variable(mod,mo, start = 14.9)

@NLexpression(mod,KR,1480.0*(3-sum{ref[1,i]*ni[i],i=1:13}))

@NLexpression(mod,Tg[j=1:size(data_fit,1)],KR/(3-sum{data_fit[j,i]*ni[i]/(data_fit[j,i]*nb_at[i]),i=1:13}))
@NLexpression(mod,mmm[j=1:size(data_fit,1)], mo * (1+sum{data_fit[j,i]*K[i],i=1:13}))

@NLexpression(mod,n_calc[j=1:size(data_fit,1)],(12-mo) + mo * Tg[j]/data_fit[j,14]*exp( (mmm[j]/mo-1)*(Tg[j]/data_fit[j,14]-1)))

@NLobjective(mod,Min,sum{(n_calc[j]-data_fit[j,15])^2,j=1:size(data_fit,1)})


ni_i = getvalue(ni)
K_i = getvalue(K)
mo_i = getvalue(mo)


KR_i = 1480.0*(3-sum(ref[1,1:13]*ni_i[1:13])) 
Tg_i = KR_i./(3-sum(data_fit[:,1:13]*ni_i[1:13]./(data_fit[:,1:13]*nb_at[1:13]),2))

M_i = mo_i .* (1+sum(data_fit[:,1:13]*K_i[1:13]))

n_pred_i = (12-mo_i) + mo_i .* Tg_i[:]./data_fit[:,14].*exp( (M_i[:]./mo_i-1).*(Tg_i[:]./data_fit[:,14]-1))

figure()
for i = 1:size(data,1)
    plot(data_fit[:,15],n_pred_i,linestyle="",marker="s",color=[data_fit[i,2]./100,data_fit[i,1]./100,(data_fit[i,3]+data_fit[i,4])./100])
end
#plot(n-no_model,Ta_model./T,linestyle="",marker="s")
```

```julia
#### Solve for the control and state
println("Solving...")
status = solve(mod)

# Display results
println("Solver status: ", status)
println("rmsd: ", getobjectivevalue(mod))
```

```julia
no_m = getvalue(no)
Ae_m = getvalue(Ae)
Be_m = getvalue(Be)
ap_m = getvalue(ap)
ScTg_m = getvalue(ScTg)
Ta_m = getvalue(Ta)

Ta_sum_mod = ones(size(data,1),4)
no_sum_mod = ones(size(data,1),4)
ap_sum_mod = ones(size(data,1),4)

for i = 1:4
    Ta_sum_mod[:,i] = (Ta_m[i,1].*(data[:,i]./100))
    no_sum_mod[:,i] = (no_m[i,1].*(data[:,i]./100))
    ap_sum_mod[:,i] = (ap_m[i,1].*(data[:,i]./100))
end

no_model = sum(no_sum_mod,2)
Ta_model = sum(Ta_sum_mod,2)
ap_model = sum(ap_sum_mod,2)

figure()
for i = 1:size(data,1)
    plot(Ta_model[i]./(T[i].*(ScTg_m+ap_model[i].*log(T[i]./Ta_model[i]))),n[i]-no_model[i],linestyle="",marker="s",color=[data[i,2]./100,data[i,1]./100,1-data[i,2]./100])
end
#plot(n-no_model,Ta_model./T,linestyle="",marker="s")
```
